### Changing colorspace

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```

```python
# another one 
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray, cmap='gray')
plt.show()
```

```python
# another one 
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.imshow(img_hsv)
plt.show()
```

```python
img_rgb.shape
```

### Resizing image 

```python
img_resize = cv2.resize(img_rgb, (500,500))
plt.imshow(img_resize)
plt.show()
```

### Blur image


```python
img_blur = cv2.blur(img_resize, (10,10))
plt.imshow(img_blur)
plt.show()
```

### Image rotation 

```python
img_rgb.shape
```

```python
row,col = img_rgb.shape[ : -1]
```

```python
Matrices = cv2.getRotationMatrix2D((row/2, col/2), -20, 0.9) #(-30) means colckwise and +(30) means anticlock wise 
```

```python
new_img = cv2.warpAffine(img_rgb, Matrices, (row, col))
plt.imshow(new_img)
plt.show()
```
