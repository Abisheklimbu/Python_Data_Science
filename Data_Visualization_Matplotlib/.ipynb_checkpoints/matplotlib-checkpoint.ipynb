### Data Visualization - Matplotlib
1. it is a library that helps us to visualize our data.
2. helps to undertands dataset in a pictorial format.
3. Extensively used in EDA(Explotary Data Analysis), and to present Machine learning model results.
4. it works best with numpy and pandas.


```python
# Install matplotlib package 
# !pip install matplotlib
```

```python
import matplotlib.pyplot as plt
import numpy as np 
```

### Line Plot

```python
# Avalable theme 
theme = plt.style.available
theme
```

```python
x = np.arange(10)
x1 = x**2
y1 = x+2
print(f'x1 ={x1}')
print(f'y1 ={y1}')
```

```python
plt.plot(x1, color = 'red', marker='o', label="Iphone")
plt.plot(y1, color='green', marker='*', markersize=10, linestyle='dashed', label="Andriod")
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Smartphone Price")
plt.legend()
plt.show()
```

### Scatter Plot
1. plt.scatter()
2. scatters data points across x and y axis.

```python
# plt.scatter?
```

```python
plt.scatter(x, x1, marker='o', label="Iphone")
plt.scatter(x, y1, marker='*', label="Andriod")
plt.xlabel("Time")
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
num = np.random.randn(100, 2)
```

```python
# [:,0] means all rows and zero index column,  num[:, 1] means all rows and one index column.
plt.scatter(num[:, 0], num[:, 1])
plt.show()
```

### Bar Graphs
1. plt.bar()
2. plot a bar graph between heights and labels you provides.

```python
x = np.arange(4)
Tesla = [40, 50, 60, 70]
Mahindra = [20, 30, 40, 50]

plt.figure(figsize=(9,6))
plt.bar(x, Tesla, tick_label = [2017, 2018, 2019, 2020], width= 0.4, label="Tesla")
plt.bar(x+0.4, Mahindra,  width= 0.4, label="Mahindra")
plt.xlabel("Years")
plt.ylabel("Sales in Millions")
plt.title("Tesla and Mahindra Car Company Sales Comparison")
plt.legend()
plt.show()
```
