### Line Plot

```python
# Avalable theme 
theme = plt.style.available
theme
```

```python
x = np.arange(10)
x1 = x**2
y1 = x+2
print(f'x1 ={x1}')
print(f'y1 ={y1}')
```

```python
plt.plot(x1, color = 'red', marker='o', label="Iphone")
plt.plot(y1, color='green', marker='*', markersize=10, linestyle='dashed', label="Andriod")
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Smartphone Price")
plt.legend()
plt.show()
```

### Bar Graphs
1. plt.bar()
2. plot a bar graph between heights and labels you provides.

```python
x = np.arange(4)
Tesla = [40, 50, 60, 70]
Mahindra = [20, 30, 40, 50]

plt.figure(figsize=(9,6))
plt.bar(x, Tesla, tick_label = [2017, 2018, 2019, 2020], width= 0.4, label="Tesla")
plt.bar(x+0.4, Mahindra,  width= 0.4, label="Mahindra")
plt.xlabel("Years")
plt.ylabel("Sales in Millions")
plt.title("Tesla and Mahindra Car Company Sales Comparison")
plt.legend()
plt.show()
```

### Subplots
1. create multiple plots in just a single figure.

```python
x = np.arange(0,10, 0.1)
y1 = x**2
y2 = x**3
y3 = np.sin(x)
y4 = np.cos(x)
```

```python
# 2 row and 2 column
plt.subplot(2,2, 1)
plt.plot(x, y1)

plt.subplot(2,2, 2)
plt.plot(x, y2)

plt.subplot(2,2, 3)
plt.plot(x, y3)

plt.subplot(2,2, 4)
plt.plot(x, y4)

plt.show()
```

```python
# importing image in array for 
img = plt.imread('luffy.jpg')
img
```

```python
# Showing image
plt.imshow(img)
# plt.axis('off')
plt.show()
```

```python
# (1227 rows, 800 column, 3 channels: RGB:Red, Green, Blue)
img.shape
```

```python
# cropping head only
new_img = img[200:600, 150:650, :]
plt.axis('off')
plt.imshow(new_img)
plt.show()
```

```python
# saving new_img 
plt.imsave('./luffy_head.jpg', new_img)
```

```python
# removing red color from original image 
img_mod = img.copy()
without_red = img_mod[:, :, 0]
plt.imshow(without_red)
plt.show()
```

```python
# now bordering original image 
bord_img = img.copy()
bord_img.shape
```

```python
rows = np.zeros((10,800,3), dtype = int)
plt.axis('off')
plt.imshow(rows)
```

```python
columns = np.zeros((1247,30,3), dtype = int)
plt.imshow(columns)
plt.axis('off')
plt.show()
```

```python
bordered_img = np.vstack((rows, bord_img, rows))
```

```python
plt.imshow(bordered_img)
plt.axis('off')
plt.show()
```

```python
boardered_img = np.hstack((columns, bordered_img, columns))
```

```python
plt.imshow(bordered_img)
plt.axis('off')
plt.show()
```

### Visualizing iris datasets

```python
import pandas as pd
iris = pd.read_csv('iris.csv')
```

```python
iris.head()
```

```python
petal_length = iris['petal_length']
petal_width = iris['petal_width']
```

```python
plt.scatter(x=petal_length, y=petal_width)
plt.xlabel('petal_length')
plt.ylabel('petal_width')
plt.show()
```

```python
setosa = iris[iris['species'] == 'setosa']
versicolor = iris[iris['species'] == 'versicolor']
virginica = iris[iris['species'] == 'virginica']
```

```python
plt.scatter(x= setosa['petal_length'], y= setosa['petal_width'], color='green')
plt.scatter(x= versicolor['petal_length'], y= versicolor['petal_width'], color='red')
plt.scatter(x= virginica['petal_length'], y= virginica['petal_width'], color='blue')
plt.xlabel('petal_length')
plt.ylabel('petal_width')
plt.show()
```

```python
# histogram 
plt.hist(iris['petal_length'], bins=20)
plt.xlabel('petal_length')
plt.ylabel('frequency')
plt.show()
```

```python
# piechart 
iris_value = iris['species'].value_counts()
```

```python
plt.pie(iris_value, autopct='%0.2f', explode = [0.1,0,0], labels=iris['species'].value_counts().index, shadow=True)
plt.show()
```

### Visualization movie datasets 

```python
movies = pd.read_csv('movie.csv')
```

```python
movies.head()
```

```python
movies.columns
```

```python
def preprocess(movie):
    return len(movie.strip())
```

```python
movies['movie_title']
```

```python
movies['movie_title_len'] = movies['movie_title' ].apply(preprocess)
```

```python
plt.hist(movies['movie_title_len'], bins = 50)
plt.xticks(range(0, 100, 5))
plt.xlabel("Movie title len" )
plt.ylabel("Number of movies (Freq)")
plt.show()
```

```python
freq_title = {}
for i in movies['movie_title_len']:
    if freq_title.get(i) is None:
        freq_title[i] =1
    else:
        freq_title[i] += 1
```

```python
x = freq_title.keys()
y = freq_title.values()
```

```python
plt.scatter(x, y)
plt.xticks(range (0, 60, 5) ) 
plt.xlabel("Movie title len" )
plt.ylabel("Number of movies (Freq)")
plt.show()
```
