```python
import matplotlib.pyplot as plt
import numpy as np 
```

### Line Plot

```python
# Avalable theme 
theme = plt.style.available
theme
```

```python
x = np.arange(10)
x1 = x**2
y1 = x+2
print(f'x1 ={x1}')
print(f'y1 ={y1}')
```

```python
plt.plot(x1, color = 'red', marker='o', label="Iphone")
plt.plot(y1, color='green', marker='*', markersize=10, linestyle='dashed', label="Andriod")
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Smartphone Price")
plt.legend()
plt.show()
```

### Scatter Plot
1. plt.scatter()

```python
# plt.scatter?
```

```python
plt.scatter(x, x1, marker='o', label="Iphone")
plt.scatter(x, y1, marker='*', label="Andriod")
plt.xlabel("Time")
plt.ylabel('Price')
plt.legend()
plt.show()
```

### Bar Graphs
1. plt.bar()
2. plot a bar graph between heights and labels you provides.

```python
x = np.arange(4)
Tesla = [40, 50, 60, 70]
Mahindra = [20, 30, 40, 50]

plt.figure(figsize=(9,6))
plt.bar(x, Tesla, tick_label = [2017, 2018, 2019, 2020], width= 0.4, label="Tesla")
plt.bar(x+0.4, Mahindra,  width= 0.4, label="Mahindra")
plt.xlabel("Years")
plt.ylabel("Sales in Millions")
plt.title("Tesla and Mahindra Car Company Sales Comparison")
plt.legend()
plt.show()
```

### Pie Chart
1. plt.pie()
2. plot the pie chart with the provides labels and values

```python
subjects = ["Math", "Science", "English", "Nepali"]
marks = [90,70,66,50]
```

```python
plt.pie(x= marks, labels=subjects, shadow=True, explode=[0.1,0,0,0], autopct="%0.1f")
plt.show()
```

### Histograms
1. plt.hist()
2. creating dummy data using normal distribution

```python
x_std = np.random.randn(600)
x_std
```

```python
math_mean = 70
math_std = 8
math_marks = np.round(x_std*math_std + math_mean)
math_marks
```

```python
plt.hist(x=math_marks, bins=10)
plt.show()
```

```python
math_marks.min()
```

```python
math_marks.max()
```

```python
# how gap is calculate through the hellp of bins
# max - min / bins , 100 -40/10 which means gap of 5 means (40-45)
```

```python
# Two histogram 
science_mean = 50
science_std = 10
science_marks = np.round(x_std*science_std + science_mean)
plt.hist(x=math_marks, bins=10)
plt.hist(x=science_marks, bins=10, alpha=0.6)
plt.show()
```

### Subplots
1. create multiple plots in just a single figure.

```python
x = np.arange(0,10, 0.1)
y1 = x**2
y2 = x**3
y3 = np.sin(x)
y4 = np.cos(x)
```

```python
# 2 row and 2 column
plt.subplot(2,2, 1)
plt.plot(x, y1)

plt.subplot(2,2, 2)
plt.plot(x, y2)

plt.subplot(2,2, 3)
plt.plot(x, y3)

plt.subplot(2,2, 4)
plt.plot(x, y4)

plt.show()
```

### 3D Plot

```python
# repeates 'a' rows upto elements of 'b'
# and repeates columns of 'b' upto elements of 'a'
x,y = np.meshgrid(x,y)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.plot_surface(x,y, x+y)
plt.show()
```
