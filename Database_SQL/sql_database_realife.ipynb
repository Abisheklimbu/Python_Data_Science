```python
import sqlite3
import pandas as pd 
```

```python
conn = sqlite3.Connection("database.sqlite")
```

```python
curr = conn.execute("SELECT * FROM Reviews LIMIT 10;")
```

```python
curr.fetchall()
```

```python
pd.read_sql_query('SELECT * FROM Reviews LIMIT 10;', conn)
```

```python
pd.read_sql_query('SELECT COUNT(*) FROM Reviews;', conn)
```

```python
pd.read_sql_query('SELECT UserId, COUNT(Score) FROM Reviews GROUP BY UserId LIMIT 10;', conn)
```

```python
pd.read_sql_query("SELECT UserId, Count (Score) count_score FROM Reviews GROUP BY UserId ORDER BY count_score DESC LIMIT 10;", conn)
```
