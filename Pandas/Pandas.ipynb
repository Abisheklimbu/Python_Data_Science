### Merging Dataframe

```python
df1 = pd.DataFrame({
    'student_name':['Ram', 'Hari', 'Shyam', 'Gita'],
    'GPA':[2,3,4,3.5]
})
df1
```

```python
df2 = pd.DataFrame({
    'teacher_name':['Ram', 'Hari', 'Shyam', 'Gita'],
    'GPA':[1,2.2,2.5,3.5]
})
df2
```

```python
# while merging two dataframe it takes same matched column name like:in this case "GPA" and their values
df1.merge(df2 , how='inner')
```

```python
# Use left side tabel and merge with right side
df1.merge(df2, how= 'left')
```

```python
# Use right side tabel and merge with left side
df1.merge(df2, how= 'right')
```

```python
# merge both: how ='left' table and how ='right' table
df1.merge(df2, how='outer')
```
