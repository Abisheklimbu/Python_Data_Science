#### Seaborn
1. seaborn is a python data visualisation libary.
2. it is built on top of matplotlib.

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
# plotting histogram plot 
sns.displot(iris['petal_length'], kind='hist')
plt.show()
```

```python
# kde plot
sns.kdeplot(iris['petal_length'])
plt.show()
```

```python
# scatter plot
sns.scatterplot(x=iris['petal_length'], y=iris['petal_width'])
plt.show()
```

```python
# scatter plot
sns.scatterplot(x='petal_length', y='petal_width', data=iris, hue='species')
plt.show()
```

```python
# jointplot 
sns.jointplot(x='petal_length', y='petal_width', data=iris, hue='species')
plt.show()
```

```python
# Pairplot
sns.pairplot(data=iris)
plt.show()
```

```python
# Pairplot
sns.pairplot(data=iris, hue='species')
plt.show()
```

```python
# Categorical plot
# countplot
sns.countplot(x='species', data=iris)
plt.show()
```

```python
# boxplot
plt.figure(figsize=(10,8))
sns.boxplot(x='species', y='petal_length', data=iris, hue='species')
plt.show()
```

```python
# violin plot
plt.figure(figsize=(10,8))
sns.violinplot(x='species', y='sepal_length', data=iris, hue='species')
plt.show()
```

```python
iris = pd.read_csv('iris.csv')
```

```python
iris
```

```python
iris.columns[4]
```

```python
species = iris.columns[4]
iris.drop(columns =[species], inplace= True)
```

```python
# correlation
iris.corr()
```

```python
# understanding correlation with scatter plot 
sns.scatterplot(x='sepal_length', y='sepal_length', data=iris)
plt.show()
```

```python
# understanding correlation with scatter plot 
sns.scatterplot(x='sepal_length', y='sepal_width', data=iris)
plt.show()
```

```python
# heatmap
plt.figure(figsize=(8,6))
sns.heatmap(data=iris.corr(), annot=True)
plt.show()
```

```python
# heatmap color options 
print(plt.colormaps)
```

```python
# heatmap
plt.figure(figsize=(8,6))
sns.heatmap(data=iris.corr(), annot=True, cmap='Reds_r')
plt.show()
```

#### Visualizing titanic datasets

```python
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.shape
```

```python
titanic.head()
```

```python
# lets count the people who survived according to passenger class
sns.countplot(x='survived', hue='pclass', data=titanic)
plt.show()
```

```python
total = titanic[titanic.columns[0]].count()
total
```

```python
# bar plot and see people who are alive according to passenger class
sns.barplot(x='survived', y='pclass' , data= titanic, hue='pclass')
plt.show()
```

```python
sns.barplot(x='alive', y='pclass', data=titanic, hue='who')
plt.show()
```

```python
sns.scatterplot(x='age', y='fare', data=titanic, hue='pclass')
plt.show()
```

```python
sns.lmplot(x='age', y='fare', data=titanic, hue='pclass')
plt.show()
```
