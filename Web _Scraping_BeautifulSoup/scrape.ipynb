```python
import bs4
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_and_their_capitals_in_native_languages'
response = requests.get(url)
response.status_code
```

```python
text = response.content
# text
```

```python
# implementing beautiful soup 
soup = bs4.BeautifulSoup(text)
soup
```

```python
# type of soup
type(soup)
```

```python
# fetching data from website
soup.title
```

```python
# if you want only title name then:
soup.title.text
```

```python
# fetching data from website first h1
soup.h1
# if you want only title name of  first h1 then:
soup.h1.text
```

```python
# Finding website tag one 
soup.find('h1')
```

```python
# Extracting table from A to Z : By using attribute class 
all_table = soup.find_all('table', attrs={"class": 'wikitable'})
```
