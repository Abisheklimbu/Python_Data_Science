#### creating csv file

```python
import pandas as pd
```

```python
print(pd.__version__)
```

```python
# blank dataframe 
df = pd.DataFrame({}, columns=['Country','Capital','Language'])
```

```python
df
```

```python
for table in all_table:
    all_tbody = table.find_all('tr')[1:]

    for tr in all_tbody:
        all_tds = tr.find_all('td')

        country = all_tds[0].text
        capital = all_tds[1].text
        language = all_tds[-1].text
        
        element = {'Country': country, 
                   'Capital':capital, 
                   'Language':language}
        
        df = df.append(element, ignore_index=True)

        print(f"{country} == {capital} == {language}")
        print()
```

```python
df
```

```python
df["Language"] = df['Language'].apply(lambda x: x[:-1])
```

```python
df
```

```python
df.head()
```

```python
# Removing spaces and []
import re
df['Country'][3]
```

```python
df["Country"] = df["Country"].apply(lambda x: re.sub('\[.*\]', "", x.strip()))
```

```python
df['Country'][3]
```

```python
# saving file into csv form
df.to_csv("country.csv", index=False)
```

### Scraping One images

```python
import bs4
import requests
```

```python
url = "https://www.passiton.com/inspirational-quotes?utf8=%E2%9C%93&q=martin+luther+king+jr"
response = requests.get(url)
responses.status_code
```

```python
content = response.content
content
```

```python
soup = bs4.BeautifulSoup(content, 'html.parser')
```

```python
all_img = soup.find_all("img")
```

```python
len(all_img)
```

```python
box = soup.find('div', attrs={"id": 'all_quotes'})
```

```python
images = box.find_all('img')
```

```python
len(images)
```

```python
img_url = images[0].attrs['src']
img_url
```

```python
res = requests.get(img_url)
res.status_code
```

```python
imgs = res.content
```

```python
with open("Quote.jpg", 'wb') as f:
    f.write(imgs)
```
